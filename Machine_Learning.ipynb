```python
import numpy as np

H = np.array([
    [2, 3-1j],
    [3+1j, 5]
])

e_vals, e_vecs = np.linalg.eigh(H)
print("Eigenvalues:", e_vals)
print("Eigenvectors:\\n", e_vecs)

psi = np.array([1/np.sqrt(2), 1/np.sqrt(2)])

probs = np.abs(e_vecs.conjugate().T @ psi)**2
print("Probabilities:", probs)
```
